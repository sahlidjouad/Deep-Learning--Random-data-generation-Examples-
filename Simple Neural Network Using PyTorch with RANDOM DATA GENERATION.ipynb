```python
import torch
import torch.nn as nn
```

```python
n_input,n_hidden,n_out,batch_size,learning_rate=10,15,1,100,0.1
```

```python
data_x=torch.randn(batch_size,n_input) #generate a random data with size batch size in n_input 
data_y=(torch.rand(size=(batch_size, 1)) < 0.5).float()

print(data_x.size())
print(data_y.size())
```

# The NN MODEL

```python
model = nn.Sequential(nn.Linear(n_input, n_hidden),
                      nn.ReLU(),
                      nn.Linear(n_hidden, n_out),
                      nn.Sigmoid())
print(model)
```

## Next, we will define the loss function and the optimizer

```python
loss_function= nn.MSELoss() # here are several different loss functions (MSE: computes the mean-squared error between the input and the target)
optimizer= torch.optim.SGD(model.parameters(),lr=learning_rate) # stochastic gradient descent (SGD) optimizer
```

# The training loop
Here, we will define the training loop with the following steps:

    Forward propagation — compute the predicted y and calculate the current loss
    Backward propagation — after each epoch we set the gradients to zero before starting to do backpropagation
    Gradient descent — Finally, we will update model parameters by calling optimizer.step() function

```python
losses = []
for epoch in range(5):
    pred_y = model(data_x)
    loss = loss_function(pred_y, data_y)
    losses.append(loss.item())

    model.zero_grad()
    loss.backward()

    optimizer.step()
```

```python
import matplotlib.pyplot as plt
plt.plot(losses)
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.title("Learning rate %f"%(learning_rate))
plt.show()
```

**We can see the loss is decreasing in each epoch which shows that the parameters are being learned.**
